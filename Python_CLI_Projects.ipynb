# Project 1: Password Strength Checker (CLI)

```python

import re
import random
import string

COMMON_PASSWORDS = ["123456", "password", "qwerty", "admin", "welcome"]

def generate_strong_password(length=8):
    chars = string.ascii_letters + string.digits + "@#$%&*!?"
    password = [
        random.choice(string.ascii_uppercase),
        random.choice(string.ascii_lowercase),
        random.choice(string.digits),
        random.choice("@#$%&*!?")
    ]
    password += [random.choice(chars) for _ in range(length - 4)]
    random.shuffle(password)
    return "".join(password)

def check_password(password):
    missing = []

    if len(password) < 8:
        missing.append("Minimum length of 8 characters")

    if not re.search(r"[A-Z]", password):
        missing.append("At least one uppercase letter")

    if not re.search(r"[a-z]", password):
        missing.append("At least one lowercase letter")

    if not re.search(r"[0-9]", password):
        missing.append("At least one number")

    if not re.search(r"[@#$%&*!?]", password):
        missing.append("At least one special character")

    if password.lower() in [p.lower() for p in COMMON_PASSWORDS]:
        print("Common password detected. Please choose a different password.")
        return

    score = 5 - len(missing)

    if score <= 2:
        strength = "Weak"
    elif score <= 4:
        strength = "Medium"
    else:
        strength = "Strong"

    print("\nPassword Strength:", strength)

    if missing:
        print("\nMissing Requirements:")
        for item in missing:
            print("-", item)

    if strength == "Weak":
        print("\nSuggestions:")
        print("- Increase password length")
        print("- Add uppercase, lowercase, numbers, and symbols")
        print("Suggested Strong Password:", generate_strong_password())

password = input("Enter Password: ")
check_password(password)
```

# Project 2: To-Do List Application (CLI)

```python

import json
import os

FILE_NAME = "tasks.txt"

def load_tasks():
    if os.path.exists(FILE_NAME):
        with open(FILE_NAME, "r") as f:
            return json.load(f)
    return []

def save_tasks(tasks):
    with open(FILE_NAME, "w") as f:
        json.dump(tasks, f)

def add_task(tasks):
    name = input("Task Name: ")
    priority = input("Priority (High/Medium/Low): ")
    tasks.append({
        "name": name,
        "status": "Pending",
        "priority": priority
    })
    save_tasks(tasks)
    print("Task Added Successfully.")

def view_tasks(tasks):
    if not tasks:
        print("No Tasks Found.")
        return

    for i, task in enumerate(tasks, start=1):
        print(f"{i}. {task['name']} | {task['status']} | {task['priority']}")

def delete_task(tasks):
    view_tasks(tasks)
    try:
        index = int(input("Enter Task Number to Delete: ")) - 1
        tasks.pop(index)
        save_tasks(tasks)
        print("Task Deleted.")
    except:
        print("Invalid Choice")

def mark_completed(tasks):
    view_tasks(tasks)
    try:
        index = int(input("Enter Task Number: ")) - 1
        tasks[index]["status"] = "Completed"
        save_tasks(tasks)
        print("Task Marked Completed.")
    except:
        print("Invalid Choice")

def search_task(tasks):
    keyword = input("Enter Keyword: ").lower()
    found = False

    for task in tasks:
        if keyword in task["name"].lower():
            print(task)
            found = True

    if not found:
        print("No Matching Task Found.")

def statistics(tasks):
    total = len(tasks)
    completed = sum(1 for t in tasks if t["status"] == "Completed")
    pending = total - completed

    print("\\nTask Statistics")
    print("Total:", total)
    print("Completed:", completed)
    print("Pending:", pending)

tasks = load_tasks()

while True:
    print("\\n===== TO-DO LIST MENU =====")
    print("1. Add Task")
    print("2. View Tasks")
    print("3. Delete Task")
    print("4. Mark Task Completed")
    print("5. Search Task")
    print("6. Task Statistics")
    print("7. Exit")

    choice = input("Enter Choice: ")

    if choice == "1":
        add_task(tasks)
    elif choice == "2":
        view_tasks(tasks)
    elif choice == "3":
        delete_task(tasks)
    elif choice == "4":
        mark_completed(tasks)
    elif choice == "5":
        search_task(tasks)
    elif choice == "6":
        statistics(tasks)
    elif choice == "7":
        print("Exiting Application...")
        break
    else:
        print("Invalid Choice")
```
