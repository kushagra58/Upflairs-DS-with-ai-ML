### 1. Explain Pandas and its role in Data Analysis and Machine Learning

**Pandas** is an open-source Python library used for data manipulation and analysis. It provides powerful data structures such as **Series** (1-dimensional) and **DataFrame** (2-dimensional) that make it easy to store, organize, clean, and analyze data.

**Role in Data Analysis:**
- Reads data from CSV, Excel, SQL, JSON, etc.
- Cleans and preprocesses data by handling missing values and duplicates.
- Filters, sorts, groups, and transforms data efficiently.
- Performs statistical analysis and generates summaries.

**Role in Machine Learning:**
- Prepares raw data before model training.
- Handles missing values and feature engineering.
- Selects and transforms required columns.
- Integrates well with libraries like NumPy, Matplotlib, Scikit-learn, and TensorFlow.

| Example: Marks of students | Example: Complete student record |
|
|

|----------|-----------|

## 3. Create a Pandas DataFrame and Perform Operations

```python
import pandas as pd

data = {
    "Name": ["Amit", "Neha", "Rahul", "Sneha"],
    "Age": [20, 21, 19, 22],
    "Marks": [85, 90, 78, 88]
}

df = pd.DataFrame(data)

print("Original DataFrame:")
print(df)

print("\nFirst 2 Rows:")
print(df.head(2))

print("\nName and Marks Columns:")
print(df[["Name", "Marks"]])

df["Result"] = ["Pass" if x >= 40 else "Fail" for x in df["Marks"]]

print("\nDataFrame with Result Column:")
print(df)

print("\nStudents with Marks Greater than 80:")
print(df[df["Marks"] > 80])
```

## 4. Handle Missing Values in a DataFrame

```python
import pandas as pd
import numpy as np

data = {
    "Name": ["A", "B", "C", "D"],
    "Age": [25, np.nan, 27, 24],
    "Salary": [30000, 40000, np.nan, 35000]
}

df = pd.DataFrame(data)

print("Original DataFrame:")
print(df)

print("\nMissing Values:")
print(df.isnull())

avg_age = df["Age"].mean()
median_salary = df["Salary"].median()

df["Age"].fillna(avg_age, inplace=True)
df["Salary"].fillna(median_salary, inplace=True)

print("\nCleaned DataFrame:")
print(df)
```

## 5. Iris Dataset Analysis

```python
import pandas as pd
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv"
iris = pd.read_csv(url)

print("First 5 Rows:")
print(iris.head())

print("\nDataset Information:")
print(iris.info())

print("\nTotal Rows and Columns:")
print(iris.shape)

print("\nAverage Sepal Length:")
print(iris["sepal_length"].mean())

print("\nDifferent Species:")
print(iris["species"].unique())

print("\nSummary Statistics:")
print(iris.describe())
```
