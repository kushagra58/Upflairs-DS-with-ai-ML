## Q1. Explain K-Nearest Neighbors (KNN)

KNN is a supervised machine learning algorithm used for **classification** and **regression**.

### Working Process
1. Choose K.
2. Compute distance (usually Euclidean).
3. Find K nearest neighbors.
4. Classification: majority vote.
5. Regression: average of neighbor values.

### Euclidean Distance
\[
d=\sqrt{\sum (x_i-y_i)^2}
\]


## Q2. Role of K

- Small K: sensitive to noise, overfitting.
- Large K: smoother decision boundary, may underfit.

Choose K using validation or odd values for binary classification.

## Q3. Classification vs Regression

|Feature|Classification|Regression|
|---|---|---|
|Output|Class label|Continuous value|
|Prediction|Majority vote|Average|

### Advantages
- Simple
- No training phase
- Non-linear boundaries

### Disadvantages
- Slow for large datasets
- Sensitive to scaling
- Memory intensive

## Q4. Student Result Prediction

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.neighbors import KNeighborsClassifier
df = pd.DataFrame({
    "Study Hours": [1, 2, 3, 4, 5, 6],
    "Attendance": [50, 55, 60, 70, 75, 85],
    "Result": ["Fail", "Fail", "Fail", "Pass", "Pass", "Pass"]
})
print("Dataset:")
print(df)
le = LabelEncoder()
df["Result"] = le.fit_transform(df["Result"])
X = df[["Study Hours", "Attendance"]]
y = df["Result"]
model = KNeighborsClassifier(n_neighbors=3)
model.fit(X, y)
new_student = pd.DataFrame({
    "Study Hours": [4.5],
    "Attendance": [72]
})
prediction = model.predict(new_student)
print("Predicted Result:", le.inverse_transform(prediction)[0])
prediction = model.predict(new_student)
print("\nStudy Hours = 4.5")
print("Attendance = 72")
print("Predicted Result =", le.inverse_transform(prediction)[0])
```

## Q5. Fruit Classification

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.neighbors import KNeighborsClassifier
fruit = pd.DataFrame({
    "Weight": [120, 130, 140, 170, 180, 200],
    "Size": [6, 6.5, 7, 8, 8.5, 9],
    "Fruit": ["Apple", "Apple", "Apple", "Orange", "Orange", "Orange"]
})
print("Dataset:")
print(fruit)
le = LabelEncoder()
fruit["Fruit"] = le.fit_transform(fruit["Fruit"])
X = fruit[["Weight", "Size"]]
y = fruit["Fruit"]
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X, y)
new_fruit = pd.DataFrame({
    "Weight": [160],
    "Size": [7.5]
})
prediction = knn.predict(new_fruit)
print("Predicted Fruit:", le.inverse_transform(prediction)[0])
prediction = knn.predict(new_fruit)
print("\nWeight = 160 g")
print("Size = 7.5 cm")
print("Predicted Fruit =", le.inverse_transform(prediction)[0])
```

## Conclusion
KNN is simple and effective for small datasets. Proper choice of K and feature scaling significantly improve performance.
