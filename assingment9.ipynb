```python

# assignment 9

# 1
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

x = [2,4,6,8,10]
y = [22,44,66,88,110]


mae = mean_absolute_error(x,y)
print("mae : ",mae)

mse = mean_squared_error(x,y)
print("mse : ",mse)

rmse = np.sqrt(mse)
print("rmse : ",rmse)

r2 = r2_score(x,y)
print("r2 :", r2)
```

```python
# # 2 .

# Evaluation Metrics Important?

# Evaluation metrics help us:

# Measure model performance.
# Compare different models.
# Detect underfitting and overfitting.
# Improve model accuracy.
# Select the best model for deployment.
```

```python
# # 3.

# Good Fit vs Overfit Models
# Metric	Good Fit Model	Overfit Model
# Training Error	Low	Very Low
# Testing Error	Low	High
# MAE	Low on both train & test	Low train, high test
# MSE	Low on both train & test	Low train, high test
# RMSE	Low on both train & test	Low train, high test
# R² Score	High on both train & test	Very high train, low test
```

```python
# 4.
X = np.array([[1],[2],[3],[4],[5]])
y = np.array([30,40,50,60,70])
```

```python

model = LinearRegression()
model.fit(X,y)
```

```python
y_pred = model.predict(X)

mae = mean_absolute_error(y,y_pred)
mse = mean_squared_error(y,y_pred)
rmse = np.sqrt(mse)

print("Predictions:", y_pred)
print("MAE =", mae)
print("MSE =", mse)
print("RMSE =", rmse)
```
