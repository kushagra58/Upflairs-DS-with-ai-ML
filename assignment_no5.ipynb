# Peterson Number


```python
def factorial(n):
    fact = 1
    for i in range(1, n + 1):
        fact *= i
    return fact

num = int(input("Enter number: "))
temp = num
total = 0

while temp > 0:
    digit = temp % 10
    total += factorial(digit)
    temp //= 10

if total == num:
    print("Peterson Number")
else:
    print("Not a Peterson Number")
```

# Circular Prime


```python
def is_prime(n):
    if n < 2:
        return False
    i = 2
    while i * i <= n:
        if n % i == 0:
            return False
        i += 1
    return True

num = input("Enter number: ")
s = num
circular_prime = True

for _ in range(len(s)):
    if not is_prime(int(s)):
        circular_prime = False
        break
    s = s[1:] + s[0]

print("Circular Prime" if circular_prime else "Not a Circular Prime")
```

# Harshad Number


```python
num = int(input("Enter number: "))
temp = num
digit_sum = 0

while temp > 0:
    digit_sum += temp % 10
    temp //= 10

if num % digit_sum == 0:
    print("Harshad Number")
else:
    print("Not a Harshad Number")
```

# Magic Number


```python
def digit_sum(n):
    total = 0
    while n > 0:
        total += n % 10
        n //= 10
    return total

num = int(input("Enter number: "))

while num > 9:
    num = digit_sum(num)

print("Magic Number" if num == 1 else "Not a Magic Number")
```

# Pascal Triangle


```python
n = int(input("Enter number of rows: "))

for i in range(n):
    num = 1
    print(" " * (n - i - 1), end="")
    for j in range(i + 1):
        print(num, end=" ")
        num = num * (i - j) // (j + 1)
    print()
```

# Duck Number


```python
num = input("Enter number: ")

duck = '0' in num[1:]

print("Duck Number" if duck else "Not a Duck Number")
```
