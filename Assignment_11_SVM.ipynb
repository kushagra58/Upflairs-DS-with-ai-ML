## Q1. Support Vector Machine (SVM)
SVM is a supervised machine learning algorithm used for classification and regression. It finds the optimal hyperplane that maximizes the margin between different classes.

### Working Principle
1. Prepare data.
2. Find the optimal hyperplane.
3. Maximize the margin.
4. Use support vectors.
5. Predict new samples.

## Q2. Support Vectors and Hyperplanes
- **Support Vectors:** Data points closest to the decision boundary.
- **Hyperplane:** A line (2D) or plane (higher dimensions) separating classes.

Example: Students classified as Pass/Fail based on study hours and attendance.

## Q3. Linear vs Non-Linear SVM and Kernel Functions
|Feature|Linear SVM|Non-Linear SVM|
|---|---|---|
|Data|Linearly separable|Non-linearly separable|
|Kernel|Linear|RBF, Polynomial, Sigmoid|

**Kernel Functions:** Transform data into a higher-dimensional space so that non-linear data can be separated.

## Q4. Student Result Prediction using SVM

```python
import pandas as pd
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

df=pd.DataFrame({
'Study Hours':[1,2,3,4,5,6],
'Attendance':[50,55,60,70,75,85],
'Result':[0,0,0,1,1,1]
})

X=df[['Study Hours','Attendance']]
y=df['Result']

model=SVC(kernel='linear')
model.fit(X,y)

new_student=pd.DataFrame({'Study Hours':[4.5],'Attendance':[72]})
prediction=model.predict(new_student)
print('Predicted Result:',prediction[0])

pred=model.predict(X)
print('Accuracy:',accuracy_score(y,pred))
df
```

## Q5. Fruit Classification using SVM

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix,classification_report

fruit=pd.DataFrame({
'Weight':[120,130,150,170,180,200],
'Size':[6,6.5,7,8,8.5,9],
'Fruit':['Apple','Apple','Apple','Orange','Orange','Orange']
})

le=LabelEncoder()
fruit['Target']=le.fit_transform(fruit['Fruit'])

X=fruit[['Weight','Size']]
y=fruit['Target']

svm=SVC(kernel='linear')
svm.fit(X,y)

new=pd.DataFrame({'Weight':[160],'Size':[7.5]})
p=svm.predict(new)
print('Predicted Fruit:',le.inverse_transform(p)[0])

yhat=svm.predict(X)
print(confusion_matrix(y,yhat))
print(classification_report(y,yhat,target_names=le.classes_))
fruit
```

## Conclusion
SVM is an effective supervised learning algorithm. Linear SVM works well for linearly separable data, while kernels enable handling non-linear problems.
